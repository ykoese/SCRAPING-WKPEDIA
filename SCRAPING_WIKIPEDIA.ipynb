```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://en.wikipedia.org/wiki/World_population"
```

```python
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data,"html5lib")
```

```python
#find all html tables in the web page
tables = soup.find_all('table') # in html table is represented by the tag <table>
```

```python
len(tables)
# 25 table var
```

```python
for index,table in enumerate(tables):
    if ("10 most densely populated countries" in str(table)):
        table_index = index
print(table_index)
# bizim aradigimiz table indexi 5
```

```python
#print(tables[table_index].prettify())
```

```python
import pandas as pd
population_data = pd.DataFrame(columns=["Rank", "Country", "Population", "Area", "Density"])

for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        rank = col[0].text
        country = col[1].text
        population = col[2].text.strip()
        area = col[3].text.strip()
        density = col[4].text.strip()
        population_data = population_data.append({"Rank":rank, "Country":country, "Population":population, "Area":area, "Density":density}, ignore_index=True)

population_data
```

```python
#Countries ranking highly in both total population 
```

```python
for index,table in enumerate(tables):
    if ("Countries ranking highly in both total population" in str(table)):
        table2_index = index
print(table2_index)
# bizim aradigimiz table indexi 5
```

```python
#print(tables[table2_index].prettify())
```

```python
population2_data = pd.DataFrame(columns=["Rank", "Country", "Population", "Area", "Density", "Population trend"])

for row in tables[table2_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        rank = col[0].text
        country = col[1].text
        population = col[2].text.strip()
        area = col[3].text.strip()
        density = col[4].text.strip()
        pop_trend=col[5].text.strip()
        population2_data = population2_data.append({"Rank":rank, "Country":country, "Population":population, "Area":area, "Density":density, "Population trend":pop_trend }, ignore_index=True)

population2_data
```
